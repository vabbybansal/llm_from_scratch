```python
import torch
```

```python
torch.manual_seed(1111)
example = torch.rand(2,5)
print(f"\nRaw example: \n{example}")

layer = torch.nn.Sequential(torch.nn.Linear(5,6), torch.nn.ReLU())
output = layer(example)
print(f"\nOutput: \n{output}")

print(f"\nOutput Mean:\n{output.mean(dim=-1)}")
print(f"\nOutput Variance:\n{output.var(dim=-1)}")
```

```python
layer_norm = torch.nn.LayerNorm(6)
output_ln = layer_norm(output)

print(f"\nLayer Norm Output Mean:\n{output_ln.mean(dim=-1)}")
print(f"\nLayer Norm Output Variance:\n{output_ln.var(dim=-1)}")
```
